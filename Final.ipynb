```python
import numpy as np

from sklearn.svm import LinearSVC

from shared.prepared_dataset import load_prepared_dataset


svc = LinearSVC()
svc.coef_ = np.load("weights.npy")
svc.intercept_ = np.load("bias.npy")
svc.classes_ = np.array([0.0, 1.0])

train_inputs, train_outputs = load_prepared_dataset("train")
test_inputs, test_outputs = load_prepared_dataset("test")
predicted_outputs = svc.predict(test_inputs)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Compute the matrix
cm = confusion_matrix(test_outputs, predicted_outputs)

# Create the display
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm, display_labels=["No Person", "Person"]
)

# Plot it
disp.plot(cmap=plt.cm.Blues)
plt.title("HOG-SVM Confusion Matrix")
plt.show()
```
